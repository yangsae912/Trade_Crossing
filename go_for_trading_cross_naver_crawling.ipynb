페이지 리스트 출력 -> 데이터 추출 -> 데이터프레임으로 데이터 넣기 -> 다음 페이지 이동 -> 페이지 리스트가 모두 돌았으면 다음 버튼 클릭 

```python
#최종 
def get_data():
    title_elements = driver.find_elements(By.CLASS_NAME, 'post_board_title__3NYcf')
    titles = [title.text for title in title_elements]
    writer_elements = driver.find_elements(By.CLASS_NAME, 'post_writer_box__2EuWj')
    writers = [x.text for x in writer_elements]
    date_elements = driver.find_elements(By.CLASS_NAME, 'post_board_information__28nF0') 
    dates = [x.text for x in date_elements]
    date = dates[::3]
    views = dates[1::3]
    likes = dates[2::3]
    
    return titles, writers, date, views, likes
```

```python
def wrt_df(titles, writers, date, views, likes):
    new_data_df = pd.DataFrame()
    new_data_df['title'] = titles
    new_data_df['writer'] = writers
    new_data_df['date'] = date
    new_data_df['views'] = views
    new_data_df['likes'] = likes
    
    return new_data_df 
```

```python
def click_next():
    next_button_elements = driver.find_elements(By.CLASS_NAME, 'article_page_move__HNBg1')
    if len(next_button_elements) != 1:
        next_button_elements[1].click()
    else:
        next_button_elements[0].click()
```

```python
# first code for page 2~ 10 (result dataframe already exists due to the first page datas)
for button_text in buttons[1:]:
    button_element = driver.find_element(By.XPATH, f"//button[text()='{button_text}']")
    button_element.click()
    time.sleep(5)
    titles, writers, date, views, likes = get_data()
    new_data_df = wrt_df(titles, writers, date, views, likes)
    result = pd.concat([result, new_data_df], ignore_index=True)
    print(f"Clicked Button: {button_text}")
    time.sleep(2)
click_next()
```

```python
button_elements = driver.find_elements(By.CLASS_NAME, 'article_page_button__2a5Iu')
buttons = [button_element.text for button_element in button_elements]
```

```python
#second code for page n1 ~ n+1 || 0 
for button_text in buttons:
    button_element = driver.find_element(By.XPATH, f"//button[text()='{button_text}']")
    button_element.click()
    time.sleep(5)
    titles, writers, date, views, likes = get_data()
    new_data_df = wrt_df(titles, writers, date, views, likes)
    result = pd.concat([result, new_data_df], ignore_index=True)
    print(f"Clicked Button: {button_text}")
    time.sleep(2)
click_next()
```

```python
while True:
    try:
        button_elements = driver.find_elements(By.CLASS_NAME, 'article_page_button__2a5Iu')
        buttons = [button_element.text for button_element in button_elements]

        for button_text in buttons:
            button_element = driver.find_element(By.XPATH, f"//button[text()='{button_text}']")
            button_element.click()
            time.sleep(5)
            titles, writers, date, views, likes = get_data()
            new_data_df = wrt_df(titles, writers, date, views, likes)
            result = pd.concat([result, new_data_df], ignore_index=True)
            print(f"Clicked Button: {button_text}")
            time.sleep(2)

        # 다음 페이지로 이동
        click_next()

    except Exception as e:
        print(f"An error occurred: {e}")
        break  # 예외가 발생하면 반복을 중지
```
